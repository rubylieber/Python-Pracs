```python
import numpy as np # This library allows us to store our data in arrays and use mathematical functions 
import matplotlib.pyplot as plt # This is the plotting package 
```

```python
# Define wavelengths
wavelength = np.arange(5,105,5)
```

```python
# Define function to calculate phase speed for a given H
def phasespeed(wavelength, H):
    g = 9.8
    kappa = 2*np.pi/wavelength
    omega = (g*kappa*np.tanh(kappa*H))**(0.5)
    C = omega/kappa
    return C
```

```python
# Calculate phase speed for a range of differnt depths
C_5 = phasespeed(wavelength, 5)
C_10 = phasespeed(wavelength, 10)
C_20 = phasespeed(wavelength, 20)
```

```python
# Plot phase speed vs wavelength for each depth 
plt.plot(wavelength, C_5, label='5m')
plt.plot(wavelength, C_10, label='10m')
plt.plot(wavelength, C_20, label='20m')
plt.legend()
plt.xlabel('Wavelength (m)')
plt.ylabel('Phase speed (m/s)')
```
