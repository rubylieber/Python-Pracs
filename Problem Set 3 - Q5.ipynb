```python
import numpy as np # This library allows us to store our data in arrays and use mathematical functions  
import matplotlib.pyplot as plt # This is the plotting package 
import cmath # This package allows us to use imaginary numbers 
```

```python
# Equivalent depth 
g = 9.8
H = 50
c = (g*H)**(0.5)
# Gradient of coriolis at the equator 
beta = 2.3*10**(-11)
# Calculate for non-dimensional wave numbers -5 to 5 
k = np.arange(-30,30.2,0.2)
# Calculate wave number in dimensional units (m-1)
# Pi defined in python as np.pi 
circ = 4*10**7
kprime = 2*np.pi*k/circ
```

```python
# Solving cubic equation, specify parameters for n=1
a1 = -(g*H) * ((kprime*kprime) + (3*beta/c))
a0 = -g*H*beta*kprime
q = a1/3
r = -(a0/2)
temp = q**3 + r*r
s1 = (r + [cmath.sqrt(t) for t in temp])**(0.33) # cmath square root function cannot be applied to array, hence need to use for loop to apply to each element 
s2 = (r - [cmath.sqrt(t) for t in temp])**(0.33)
```

```python
# First solution 
sigma1_orig = (s1+s2)
# Second solution (gives only negative frequencies) 
sigma2_orig = -0.5*(s1+s2)+0.866j*(s1-s2) # j denotes imaginary unit in python 
# Third solution 
sigma3_orig = -0.5*(s1+s2)-0.866j*(s1-s2) 
```

```python
# Now for n=-1
# First solution (kelvin wave) 
sigma = c*kprime
```

```python
# Plot all solutions  
plt.plot(kprime, sigma1_orig, c='green')
plt.plot(kprime, sigma2_orig, c='red')
plt.plot(kprime, sigma3_orig, c='orange')
plt.plot(kprime,sigma, c='black', linestyle='--')
plt.ylim(0, 10**(-4))
plt.xlabel('Zonal Wavenumber (m-1)')
plt.ylabel('Frequency (s-1)')
```

```python
# Scaling
k_scale = kprime*(c/beta)**0.5
sigma_scale1 = sigma1_orig/(beta*c)**0.5
sigma_scale2 = sigma2_orig/(beta*c)**0.5
sigma_scale3 = sigma3_orig/(beta*c)**0.5
sigma_kscale = c*kprime/(beta*c)**0.5
```

```python
# Plot scaled solutions
plt.plot(k_scale, sigma_scale1, c='green')
plt.plot(k_scale, sigma_scale2, c='red')
plt.plot(k_scale, sigma_scale3, c='orange')
plt.plot(k_scale, sigma_kscale, c='black', linestyle='--')
plt.ylim(0, 4)
plt.xlabel('Scaled Zonal Wavenumber')
plt.ylabel('Scaled Frequency')
```
